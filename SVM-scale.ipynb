```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('result.xlsx')[0:32]
data
```

```python
X=data.iloc[:,1:4]
```

```python
X=np.array(X)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))

# Fit and transform the data
X_scale = scaler.fit_transform(X)
```

```python
X_scale 
```

```python
y=data[['phase/ 0 alpha/1 beta/2 mix']]
```

```python
y=np.array(y).ravel()
```

```python
y
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
```

```python
clf_scaled = make_pipeline(MinMaxScaler(), SVC(kernel='rbf', probability=True, random_state=42))
scores_scaled = cross_val_score(clf_scaled, X, y, cv=5,scoring='accuracy')
mean_cv_acc = scores_scaled.mean()
std_cv_acc = scores_scaled.std()
sem_cv_acc = std_cv_acc / np.sqrt(len(scores_scaled))
print(mean_cv_acc, sem_cv_acc)
```
